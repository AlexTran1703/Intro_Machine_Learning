```python
#Lab 1 Problem 1
import numpy as np
import matplotlib.pyplot as plt

def zero_to_nan(array):
    """ Return nan for any value = 0"""
    return [float('nan') if x==0 else x for x in array]

def find_vertex(arr):
  """Return array with vertexs of the input array"""
  #Size of the vertex array = sizr of input array
  arr_v = np.zeros((arr.shape[0],), dtype = int)
  for i in range(arr.shape[0]):
    #Index j and k for finding vertex
    if i<arr.shape[0]-1:
      j = i + 1
    else:
      j = i
    if i>0:
      k = i - 1
    else:
      k = i
    #Finding vertexes of array
    #if i==0:
      #if arr[i] > arr[j]:
        #arr_v[i] = arr[i]   
    #if i==arr.shape[0]-1:
      #if arr[i] > arr[k]:
        #arr_v[i] = arr[i]
    if i>0 & i<arr.shape[0]-1:
      if (arr[i] > arr[j]) & (arr[i] > arr[k]):
        arr_v[i] = arr[i]

  #any values = 0 will be nan
  arr_v = zero_to_nan(arr_v)
  return arr_v

#Create 4 arrays
b1 = np.array([1, 3, 7, 1, 2, 6, 0, 1])
b2 = np.array([4, 6, 10, 11, 4, 2, 7, 4, 2])
b3 = np.array([7, 4, 5, 4, 6, 8, 7, 3])
b4 = np.array([5, 3 ,4 ,6 ,4, 6, 7])

b_ran = np.random.random(12)

#Create 4 arrays with vertexs
c1 = find_vertex(b1)
c2 = find_vertex(b2)
c3 = find_vertex(b3)
c4 = find_vertex(b4)

#Plot arrays
figure, axis = plt.subplots(2, 2)
axis[0, 0].plot(b1, 'b', c1, 'r.')
axis[0, 1].plot(b2, 'b', c2, 'r.')
axis[1, 0].plot(b3, 'b', c3, 'r.')
axis[1, 1].plot(b4, 'b', c4, 'r.')
plt.show()
```
