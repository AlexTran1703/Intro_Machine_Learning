```python
#Lab 1 Problem 2.b
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import drive 
drive.mount('/content/drive/')

image_path = '/content/drive/MyDrive/Colab Notebooks/Colab_data/mrbean.jpg'
#Read image 
image = cv2.imread(image_path,0)

def conv_2d(img, size_ker):
    """Convolution 2D accept 2D image and kernel size"""
    #Create blur kernel
    kernel = np.ones((size_ker,size_ker), np.float32) / float(size_ker**2)
    
    #Read size of kernel and image
    x_ker, y_ker = kernel.shape
    x_img, y_img = image.shape

    #Create size of filter image
    #Due to padding, the size will be reduced depending on kernel size
    x_img = x_img - x_ker + 1
    y_img = y_img - y_ker + 1
    
    #Create zeros matrix
    fil_img = np.zeros((x_img,y_img))

    #Convolution step
    for i in range(x_img):
      for j in range(y_img):
        fil_img[i][j] = np.sum(image[i:i+x_ker, j:j+y_ker]*kernel)

    return fil_img

#Create 2 filter image with size 3 and 6 and 12
image_ker3 = conv_2d(image,3)
image_ker6 = conv_2d(image,6)

print("\nSize of Original Image")
print(f"{image.shape}")
print("\nSize of filter image Kernel 3")
print(f"{image_ker3.shape}")
print("\nSize of filter image Kernel 6")
print(f"{image_ker6.shape}\n\n\n")


cv2_imshow(image)
cv2_imshow(image_ker3)
cv2_imshow(image_ker6)


```
