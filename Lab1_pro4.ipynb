```python
#Lab 1 Problem 4
import cv2
import random
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import drive 
drive.mount('/content/drive/')

image_path = '/content/drive/MyDrive/Colab Notebooks/Colab_data/mrbean2.jpg'
face_reg = cv2.CascadeClassifier('/content/drive/MyDrive/Colab Notebooks/Colab_data/haarcascade_frontalface_default.xml')
#Read image 
image = cv2.imread(image_path)

def face_recognition(image):
  """This function is for face detection of an specific image"""
  #Get a copy of input image
  new_image = image.copy()
  # Convert into grayscale
  gray = cv2.cvtColor(new_image, cv2.COLOR_BGR2GRAY)
  # Detect faces
  faces_crop = face_reg.detectMultiScale(gray, 1.3, 5)
  # Draw rectangle around the faces
  for (x, y, w, h) in faces_crop:
    cv2.rectangle(new_image, (x, y), (x+w, y+h), (255, 0, 0), 2)
    #Display face dectection image
    cv2_imshow(new_image)

    #Face crop from the face dectection image
    faces_crop = new_image[y:y + h, x:x + w]

    #Copy a copy of face dectection image to use for bluring
    blur_image = new_image.copy()
    blur_image[y:y + h, x:x + w] = cv2.medianBlur(faces_crop, 5)

    #Display blur face image
    cv2_imshow(blur_image)

#cv2_imshow(image)
face_recognition(image)




```
