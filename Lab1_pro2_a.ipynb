```python
#Lab 1 Problem 2.a
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import drive 
drive.mount('/content/drive/')

image_path = '/content/drive/MyDrive/Colab Notebooks/Colab_data/mrbean.jpg'
#Read image 
image = cv2.imread(image_path)

#print(image.ndim)

# 4 size of kernel
size_ker1 = 5
size_ker2 = 10
size_ker3 = 15
size_ker4 = 20

#Create 4 differences kernel
kernel1 = np.ones((size_ker1,size_ker1), np.float32) / float(size_ker1**2)
kernel2 = np.ones((size_ker2,size_ker2), np.float32) / float(size_ker2**2)
kernel3 = np.ones((size_ker3,size_ker3), np.float32) / float(size_ker3**2)
kernel4 = np.ones((size_ker4,size_ker4), np.float32) / float(size_ker4**2)

# 4 different filtered images
image_fil1 = cv2.filter2D(image, cv2.CV_8U, kernel1)
image_fil2 = cv2.filter2D(image, cv2.CV_8U, kernel2)
image_fil3 = cv2.filter2D(image, cv2.CV_8U, kernel3)
image_fil4 = cv2.filter2D(image, cv2.CV_8U, kernel4)

#print(image_fil1[0][:])

#stack 4 filtered images in horizontal line
image_stack = np.concatenate((image_fil1, image_fil2, image_fil3, image_fil4), axis=1)

#Display image
cv2_imshow(image_stack)
cv2.waitKey()
```
