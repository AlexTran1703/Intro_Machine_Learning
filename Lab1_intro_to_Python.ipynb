````python
def fac(n):
  if n == 1:
    return 1
  else:
    return n*fac(n-1)

def expon(n):
  if n == 0:
    return 1
  else:
    return 1/(fac(n)) + expon(n-1)

n = float(input("n = "))

print(f"exp = {expon(n)}")

````

# MEDIAN

````python
def calculate_median(numbers):
  N = len(numbers)
  numbers.sort()

  if N%2 == 0:
    m1 = N/2
    m2 = (N/2) + 1
    m1 = int(m1) - 1
    m2 = int(m2) - 1
    median = float(numbers[m1] + numbers[m2]) / 2
  else:
    m = (N + 1)/2
    m = int(m) - 1
    median = numbers[m]
  return median 
````
